# **Bar Graph**

### **1. Create a vertical bar chart to compare sales of 5 different products.Add axis labels, title, and customize bar colors.**

```python
import matplotlib.pyplot as plt

products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [150, 200, 120, 300, 250]

colors = ['red', 'blue', 'purple', 'green', 'orange']

plt.bar(products, sales, color=colors)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Sales of Products')
plt.show()
```

### **2. Plot a horizontal bar chart showing the number of students enrolled in different courses (Python, ML, AI, DSA).**

```python
courses = ['Python', 'ML', 'AI', 'DSA']
students = [30, 25, 20, 22]
colors = ['Black', 'Green', 'Blue', 'Red']
plt.barh(courses, students, color=colors)
plt.xlabel('Number of Students')
plt.ylabel('Courses')
plt.title('Number of Students Enrolled in Different Courses')
plt.show()
```

### **3. Create a grouped bar chart comparing scores of two students across 5 subjects.**

```python
import numpy as np

# Subjects and scores
subjects = ['Math', 'Science', 'English', 'History', 'Art']
student1_scores = [85, 90, 78, 88, 76]
student2_scores = [80, 85, 82, 75, 89]

# Set bar width and positions
bar_width = 0.35
x = np.arange(len(subjects))

# craete grouped bars
plt.bar(x - bar_width/2, student1_scores, bar_width, label='Student 1', color='blue')
plt.bar(x + bar_width/2, student2_scores, bar_width, label='Student 2', color='green')

# Add labels and title
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.title('Scores of Two Students Across Subjects')
plt.xticks(x, subjects)
plt.legend()

# Show the plot
plt.show()
```

### **4. Plot a stacked bar chart showing monthly expenses split into categories: rent, food, transport, and savings.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Months and categories
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
rent = np.array([800, 800, 800, 800, 800])
food = np.array([300, 320, 280, 310, 290])
transport = np.array([100, 120, 110, 90, 105])
savings = np.array([200, 220, 210, 250, 240])

# bar positions
x = np.arange(len(months))

# plot stacked bars
plt.bar(x, rent, label='Rent', color='blue')
plt.bar(x, food, bottom=rent, label='Food', color='green')
plt.bar(x, transport, bottom=rent+food, label='Transport', color='red')
plt.bar(x, savings, bottom=rent+food+transport, label='Savings', color='orange')

plt.xlabel('Months')
plt.ylabel('Expenses')
plt.title('Monthly Expenses')
plt.xticks(x, months)
plt.legend()
plt.show()
```

# **Histogram**

### **5. Generate a histogram showing the distribution of student ages (randomly generated between 18–30). Use different bins values and compare results.**

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42) # for reproducability
ages = np.random.randint(18, 31, 100)
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
# histogram with fewer bins
plt.hist(ages, bins=5, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram with 5 bins')
# histogram with more bins
plt.subplot(1, 2, 2)
plt.hist(ages, bins=20, color='lightgreen', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram with 20 bins')
plt.tight_layout()
plt.show()
```

### **6. Plot a histogram of normally distributed data (mean=50, std=10) and highlight the mean using a vertical line.**

```python
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000) # mean=50, std=10

plt.hist(data, bins=30, color='red', edgecolor='black', alpha=0.5)

mean_value = np.mean(data)
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {mean_value:.2f}')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Data with Mean Line')
plt.legend()
plt.show()
```

# **Piechart**

### **8. Create a pie chart showing market share of 4 tech companies (Apple, Google, Amazon, Microsoft). Add percentage labels.**

```python
tech_comp = ['Apple', 'Google', 'Amazon', 'Microsoft']
market_share = [0.4, 0.3, 0.2, 0.1]
colors = ['lightpink', 'lightblue', 'lightgreen', 'orange']
plt.pie(market_share, labels=tech_comp, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Market Share of Tech Companies')
plt.show()
```

### **9. Make a pie chart showing daily activities (work, sleep, exercise, leisure). Explode one slice to emphasize the maximum time spent activity.**

```python
activities = ['Work', 'Sleep', 'Exercise', 'Leisure']
time_spent = [8, 7, 2, 7]  # hours
# Find index of maximum time activity
max_index = time_spent.index(max(time_spent))
# Explode only the maximum slice
explode = [0.1 if i == max_index else 0 for i in range(len(activities))]
# Plot pie chart
plt.pie(time_spent, labels=activities, autopct='%1.1f%%',
        startangle=90, explode=explode, colors=['red', 'lightgreen', 'orange', 'pink'])
plt.title('Daily Activities Breakdown')
plt.show()
```

### **10.	Create a donut chart (pie chart with a hole) showing the proportion of machine learning model usage (Regression, Classification, Clustering, Others).**

```python
models = ['Regression', 'Classification', 'Clustering', 'Others']
usage = [30, 45, 15, 10]  # percentages

# Plot pie chart
plt.pie(usage, labels=models, autopct='%1.1f%%', startangle=90,
        colors=['skyblue', 'salmon', 'lightgreen', 'orange'], wedgeprops={'linewidth': 1, 'edgecolor': 'white'})

# Add a circle at the center to make it a donut
centre_circle = plt.Circle((0,0), 0.30, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Title
plt.title('Machine Learning Model Usage (Donut Chart)')

plt.show()
```

### **12. Generate a scatter plot of height vs weight, where marker size depends on age and color depends on gender.**

```python
height = [150, 160, 165, 170, 175, 180, 185, 190]
weight = [50, 55, 60, 65, 70, 75, 85, 90]
age =    [18, 20, 22, 25, 28, 30, 35, 40]   # marker size
gender = ['F', 'M', 'F', 'M', 'F', 'M', 'F', 'M']  # marker color

# Map genders to colors
colors = ['pink' if g == 'F' else 'blue' for g in gender]

# Create scatter plot
plt.scatter(height, weight, s=[a*10 for a in age], c=colors, alpha=0.6, edgecolor='black')

# Labels and title
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight (Size = Age, Color = Gender)')

# Custom legend
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Female', markerfacecolor='pink', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Male', markerfacecolor='blue', markersize=10)
]
plt.legend(handles=legend_elements, title='Gender')

plt.show()
```

### **13. Visualize Iris dataset (sepal length vs sepal width) using plt.scatter with different colors for each species.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target
species = iris.target_names

# Sepal length = column 0, Sepal width = column 1
sepal_length = X[:, 0]
sepal_width = X[:, 1]

# Colors for species
colors = ['red', 'green', 'blue']

# Scatter plot
for i, color in enumerate(colors):
    plt.scatter(sepal_length[y == i], sepal_width[y == i],
                color=color, label=species[i], alpha=0.7, edgecolor='black')

# Labels and title
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Dataset: Sepal Length vs Sepal Width')
plt.legend()

plt.show()
```

# **Customization**

###**14. Draw a scatter plot with different marker styles (o, x, ^, *) and line styles (--, :, -.) for multiple data series.**

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]

y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [2, 3, 4, 5, 6]
y4 = [5, 4, 3, 2, 1]

# Plot each series with different marker & line style
plt.plot(x, y1, marker='o', linestyle='--', color='blue', label='Series 1')
plt.plot(x, y2, marker='x', linestyle=':', color='red', label='Series 2')
plt.plot(x, y3, marker='^', linestyle='-.', color='green', label='Series 3')
plt.plot(x, y4, marker='*', linestyle='-', color='purple', label='Series 4')

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Different Marker & Line Styles')
plt.legend()

plt.show()
```
